```
import urllib.request
import json

# please pass in a different url if necessary
def get_api_information(api_keywords, symbol, api_token, url="https://cloud.iexapis.com/stable/"):
  for keyword in api_keywords:
    print("\n")
    print("--------------------------------------------------------------")
    print(f"Return for call: \"{keyword}\"")
    print("--------------------------------------------------------------")
    if '{' in keyword:
      keyword = keyword.replace('{symbol}', symbol)
    try:
      contents = urllib.request.urlopen(f"{url}{keyword}?token={api_token}")

      response_body = contents.read().decode('utf-8')
      response_format = json.loads(response_body)
      json_formatted_response = json.dumps(response_format, indent=2)
      print(json_formatted_response)
      print("------------------------------------------------------------")
    except Exception as exc:
      logger.warning("error while processing item: %s", exc)
      logger.warning(f"Occured with API call: \"{keyword}\"")
      continue
```
